```python
%sql
CREATE TEMP VIEW tempview2_bike AS
SELECT 
  month(dteday) as mes,
  avg(cnt) as media_ctn
FROM lab_certif.aula01.bikes_bronze
GROUP BY mes
ORDER BY mes;
```

```python
%sql
SELECT * FROM tempview2_bike
```
