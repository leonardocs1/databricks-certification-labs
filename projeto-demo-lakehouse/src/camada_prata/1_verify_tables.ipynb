## ✅ Verificação das tabelas da camada Silver

Este notebook verifica a existência das tabelas da camada `adventure_works_silver`
e imprime a contagem de linhas para confirmar se estão disponíveis antes de iniciar a camada ouro.

---

```python
from pyspark.sql.utils import AnalysisException

# Lista de tabelas da camada prata
silver_tables = [
    "person_address",
    "person_countryregion",
    "person_emailaddress",
    "person_person",
    "person_personphone",
    "person_stateprovince",
    "production_product",
    "production_productcategory",
    "production_productdescription",
    "production_productmodel",
    "production_productsubcategory",
    "sales_currency",
    "sales_customer",
    "sales_salesorderdetail",
    "sales_salesorderheader",
    "sales_salesorderheadersalesreason",
    "sales_salesreason",
    "sales_salesterritory",
    "sales_specialoffer"
]

# Loop para verificar cada tabela
for table in silver_tables:
    full_name = f"adventure_works_silver.{table}"
    try:
        df = spark.read.table(full_name)
        count = df.count()
        print(f"✅ Tabela encontrada: {full_name} | Linhas: {count}")
    except AnalysisException:
        print(f"❌ Tabela não encontrada: {full_name}")
```
