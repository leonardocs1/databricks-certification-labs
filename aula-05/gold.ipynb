```python
%sql
CREATE OR REPLACE TABLE formula1.gold.points_by_driver
SELECT
  driverRef,
  --constructorRef,
  SUM(points) AS points_driver
FROM
  formula1.bronze.results res
    INNER JOIN formula1.bronze.drivers dr
      ON res.driverId = dr.driverId
    INNER JOIN formula1.bronze.constructors ct
      ON res.constructorId = ct.constructorId
GROUP BY
  driverRef
ORDER BY
  SUM(points) DESC
```
