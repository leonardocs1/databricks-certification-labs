```python
schema_loc = "/Volumes/formula1/landing/formula1_autoloader/_schemas/laps"
checkpoint = "/Volumes/formula1/landing/formula1_autoloader/_checkpoints/laps"

df = (spark.readStream
      .format("cloudFiles")
      .option("cloudFiles.format", "parquet")
      .option("cloudFiles.schemaLocation", schema_loc)
      .option("cloudFiles.schemaEvolutionMode", "addNewColumns")
      .load("/Volumes/formula1/landing/formula1_autoloader/") 
)
```

```python
(df.writeStream
   .format("delta")
   .option("checkpointLocation", checkpoint)
   .outputMode("append")
   .trigger(availableNow=True)
   .table("formula1.bronze.lap_times_autoloader")
)
```
