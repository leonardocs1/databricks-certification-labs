```python
%sql
USE CATALOG formula1;
CREATE VOLUME IF NOT EXISTS formula1.landing.formula1_autoloader;
```

```python
%sql
SELECT * FROM formula1.bronze.lap_times
```

```python
%sql
SELECT * FROM formula1.bronze.races;
```

```python
year = 2008

spark.sql(
    f"""
        SELECT * FROM formula1.bronze.lap_times WHERE raceId IN (
            SELECT raceId FROM formula1.bronze.races WHERE year = {year}
        )
    """
).write.format("parquet").mode("overwrite").save(
    f"/Volumes/formula1/landing/formula1_autoloader/laps{year}.parquet"
)
```

```python
for year in range(2005, 2006):
    spark.sql(
        f"""
            SELECT * FROM formula1.bronze.lap_times WHERE raceId IN (
                SELECT raceId FROM formula1.bronze.races WHERE year = {year}
            )
        """
    ).write.format("parquet").mode("overwrite").save(
        f"/Volumes/formula1/landing/formula1_autoloader/laps{year}.parquet"
    )
```
